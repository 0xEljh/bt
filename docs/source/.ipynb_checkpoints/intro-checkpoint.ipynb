### A Simple Strategy Backtest

Let's create a simple strategy. We will create a monthly rebalanced, long-only strategy where we place equal weights on each asset in our universe of assets.

Now we should download some data. By default, **bt.get** downloads the Adjusted Close from Yahoo! Finance. We will download some data starting on January 1, 2010 for the purposes of this demo.

We will also have to create a **Backtest**, which is the logical combination of a strategy with a data set.

Once this is done, we can run the backtest and analyze the results.

```python
# create the strategy
s = bt.Strategy('s1', [bt.algos.RunMonthly(),
                       bt.algos.SelectAll(),
                       bt.algos.WeighEqually(),
                       bt.algos.Rebalance()])

# fetch some data
data = bt.get('spy,agg', start='2010-01-01')

# create a backtest and run it
test = bt.Backtest(s, data)
res = bt.run(test)
```

Now we can analyze the results of our backtest. The **Result** object is a thin wrapper around **ffn.GroupStats** and adds some helper methods.

```python
# first let's see an equity curve
res.plot()
```

```python
# ok and what about some stats?
res.display()
```

```python
# ok and how does the return distribution look like?
res.plot_histogram()
```

```python
# and just to make sure everything went along as planned, let's plot the security weights over time
res.plot_security_weights()
```

Now what if we ran this strategy weekly and also used some risk parity style approach by using weights that are proportional to the inverse of each asset's volatility? Well, all we have to do is plug in some different algos. See below:

```python
# create our new strategy
s2 = bt.Strategy('s2', [bt.algos.RunWeekly(),
                        bt.algos.SelectAll(),
                        bt.algos.WeighInvVol(),
                        bt.algos.Rebalance()])
```

```python
# now let's test it with the same data set. We will also compare it with our first backtest.
test2 = bt.Backtest(s2, data)
res2 = bt.run(test, test2)
```

```python
res2.plot()
```
